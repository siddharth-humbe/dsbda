```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].value_counts()
```

```python
df['Age']=df['Age'].astype(int)
df.Age.value_counts()
```

```python
# Calculate the minimum and maximum age
min_age = df['Age'].min()
max_age = df['Age'].max()

# Calculate the number of bins based on the range and desired interval size
interval_size = 5 # or 5
num_bins = int((max_age - min_age) / interval_size) + 1

# Generate bin edges
bins = [min_age + i * interval_size for i in range(num_bins)]
bins.append(max_age)  # Add the maximum age to the last bin

# Generate labels
labels = [f'{bins[i]}-{bins[i+1]-1}' for i in range(len(bins)-2)]
labels.append(f'{bins[len(bins)-2]} and above' )

labels
# Perform binning
# df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df['AgeGroup']= pd.cut(df['Age'], bins= bins, labels= labels, right= True)
```
