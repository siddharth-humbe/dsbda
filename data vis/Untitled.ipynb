```python
mx = df.fare.max()
mn = df.fare.min()

binsize = 100

noofbins = int((mx-mn)/binsize) + 1
bins = [mn + i*binsize for i in range(noofbins) ]

labels= [f'{bins[i]}-{bins[i+1]-1}' for i in range(noofbins-1)]
labels.append('above 500')
```

```python
labels
```

```python
sns.histplot(data= df, x='fare', bins=bins)
```
