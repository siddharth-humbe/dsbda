```python
alphabet = string.printable
```

```python
def clean_doc(doc):
    return ''.join([c for c in doc.lower() if c in alphabet])

doc1= clean_doc(doc_01)
doc2= clean_doc(doc_02)
```

stop word removal

```python
def remove_stopword(tokens):
    return [token for token in tokens if token not in nltk.corpus.stopwords.words('english')]

stopremove_1= remove_stopword(token1)
```

```python
def TF(tokens):
    num_tokens= len(tokens)
    unique_tokens, frequency= np.unique(tokens, return_counts= True)
    tf={}
    for token, freq in zip(unique_tokens, frequency):
        tf[token]= freq/num_tokens
    return tf

def IDF():
    N=2
    total_tokens= lemmet_token1+lemmet_token2
    
    idf={}
    
    for token in total_tokens:
        f=0
        if(token in lemmet_token1):
            f+=1
        if(token in lemmet_token2):
            f+=1
        
        idf[token]= math.log10(N/f)
        
    return idf



tf1= TF(lemmet_token1)
tf2= TF(lemmet_token2)
idf = IDF()
```

```python
doc_1_repr = []
for token in lemmet_token1:
    doc_1_repr.append( tf1[ token ] * idf[token] )
doc_2_repr = []
for token in lemmet_token2:
    doc_2_repr.append( tf2[ token ] * idf[token] )
```
