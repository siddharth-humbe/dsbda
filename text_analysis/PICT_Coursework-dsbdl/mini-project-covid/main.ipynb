#### Visualizations for no. of doses, sites, sessions

````python
ax = sns.lineplot( data=ds, x="Updated On", y="Total Doses Administered" )
ax.set(xticklabels=[])
ax.tick_params(bottom=False)
plt.show()
````

````python
ax = sns.lineplot( data=ds, x="Updated On", y="Sessions" )
ax.set(xticklabels=[])
ax.tick_params(bottom=False)
plt.show()
````

````python
ax = sns.lineplot( data=ds, x="Updated On", y=" Sites " )
ax.set(xticklabels=[])
ax.tick_params(bottom=False)
plt.show()
````

````python
ax = sns.lineplot( data=ds[[ "Updated On", "First Dose Administered","Second Dose Administered"]]  )
ax.set(xticklabels=[])
ax.tick_params(bottom=False)
plt.show()
````

### First Dose Administered - aggregated state wise

````python
first_dose_ds = ds.loc[ (ds[ "First Dose Administered" ] > 0) & (ds["State"] != "India") ]
first_dose_ds = first_dose_ds[ [ "State" , "First Dose Administered" ]  ]
first_dose_ds = first_dose_ds.groupby( "State" ).agg( 'sum' )
first_dose_ds.head()
````

````python
ax = sns.barplot( data=first_dose_ds, x="State", y="First Dose Administered" )
ax.tick_params(axis='x', rotation=90)
ax.tick_params(bottom=False)
plt.show()
````

### Second Dose Administered - aggregated state wise

````python
second_dose_ds = ds.loc[ (ds[ "Second Dose Administered" ] > 0) & (ds["State"] != "India") ]
second_dose_ds = second_dose_ds[ [ "State" , "Second Dose Administered" ]  ]
second_dose_ds = second_dose_ds.groupby( "State" ).agg( 'sum' )
second_dose_ds.head()
````

````python
ax = sns.barplot( data=second_dose_ds, x="State", y="Second Dose Administered" )
ax.tick_params(axis='x', rotation=90)
ax.tick_params(bottom=False)
plt.show()
````

### Number of males vaccinated - aggregated state wise

````python
sub = ds.loc[ (ds[ "Male (Doses Administered)" ] > 0) & (ds["State"] != "India") ]
sub = sub[ [ "State" , "Male (Doses Administered)" ]  ]
sub = sub.groupby( "State" ).agg( 'sum' )
sub.head()
````

````python
ax = sns.barplot( data=sub, x="State", y="Male (Doses Administered)" )
ax.tick_params(axis='x', rotation=90)
ax.tick_params(bottom=False)
plt.show()
````

### Number of females vaccinated - aggregated state wise

````python
sub = ds.loc[ (ds[ "Female (Doses Administered)" ] > 0) & (ds["State"] != "India") ]
sub = sub[ [ "State" , "Female (Doses Administered)" ]  ]
sub = sub.groupby( "State" ).agg( 'sum' )
sub.head()
````

````python
ax = sns.barplot( data=sub, x="State", y="Female (Doses Administered)" )
ax.tick_params(axis='x', rotation=90)
ax.tick_params(bottom=False)
plt.show()
````
