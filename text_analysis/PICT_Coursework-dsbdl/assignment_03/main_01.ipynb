```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv( "winequality-red.csv" , sep=";" )
ds.head()
```

## 3.1. Summary-statistics of numeric features

```python
ds.drop( [ "quality" ] , axis=1 ).describe()
```

```python
numeric_features = [ "fixed acidity" ,
"citric acid" ,
"residual sugar" ,
"chlorides" ,
"free sulfur dioxide" ,
"total sulfur dioxide" ,
"density" ,
"pH" ,
"sulphates" ,
"alcohol" ]
```

```python
fig, axes = plt.subplots(5 , 2 , figsize=( 10 , 10 ))
axes = axes.flatten()
for i , feature in enumerate( numeric_features ):
    sns.boxplot( data=ds , y=feature, x="quality" , ax=axes[i] )
```

### 3.1. Summary-statistics of numeric features grouped by `quality`

```python
def group_stats( feature_name ):
    labels = ds[ "quality" ].unique().tolist()
    for label in labels:
        print( f"Label: {label}")
        print( ds[ ds[ "quality" ] == label ][ feature_name ].describe() , end="\n\n" )
```

```python
for feature in numeric_features:
    group_stats( feature )
```
