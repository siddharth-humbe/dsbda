#Document PreProcessing

Tokenization

```python
token1 = nltk.word_tokenize(doc1)
token2 = nltk.word_tokenize(doc2)
print(token1)
```

pos tagging

```python
pos_tag = dict( zip( token1, nltk.pos_tag(token1) ) )
pos_tag
```

sentence tokenization

```python
sent_token1 = nltk.sent_tokenize(doc1)
sent_token2= nltk.sent_tokenize(doc2)
```

lemmatization

```python
lemmatizer = nltk.stem.WordNetLemmatizer()

token1 = [lemmatizer.lemmatize(token,pos='v') for token in token1]

token2 = [lemmatizer.lemmatize(token, pos='v') for token in token2]


print(token1)
# token2 = [lemmatizer.lemmatize(token) for token in token2]

# print(token1)
```

Stemming

```python
stemmer = nltk.stem.PorterStemmer()
token1 = [stemmer.stem(token) for token in token1]
token2 = [stemmer.stem(token) for token in token2]
print(token1)
```

stopword removal

```python
def remove_stop_words(tokens):
    return [token for token in tokens if token not in nltk.corpus.stopwords.words('english')]


token1 = remove_stop_words(token1)
token2= remove_stop_words(token2)
print(token1)
```

TF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
def TF(word_tokens):
    num_tokens= len(word_tokens)
    unique_tokens, frequency= np.unique(word_tokens, return_counts= True)
    tf= {}
    for token, freq in zip(unique_tokens, frequency):
        tf[token] = freq/num_tokens
    return tf

# def TF( doc_tokens ):
#     N = len( doc_tokens )
#     token_freq = dict( [ ( token , 0 ) for token in doc_tokens ] )
#     for token in doc_tokens:
#         token_freq[ token ] += 1
#     tf = dict( [ ( token , count / N ) for token , count in token_freq.items() ] )
#     return tf

```

```python
token_1 = token1.copy()
token_2 = token2.copy()
```

```python
def IDF():
    N=2
    all_tokens= token_1+ token_2
    
    idf= {}
    
    for token in all_tokens:
        f=0
        if token in token_1:
            f+=1
        if token in token_2:
            f+=1
        idf[token] = math.log(N/f)
    return idf
```

```python
tf1 = TF(token_1)
tf2 = TF(token_2)
idf = IDF()
```

```python
doc_1_repr = []
for token in token_1:
    doc_1_repr.append( tf1[ token ] * idf[token] )
doc_2_repr = []
for token in token_2:
    doc_2_repr.append( tf2[ token ] * idf[token] )
```
